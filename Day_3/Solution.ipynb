# Part 1

```python
# counts trees with right 3, down 1
count = 0
tree = 0

for y in df.index:
    
    if count % 3 == 0:
        
        if df[count][y] == 1:
            
            tree +=1
            
    count+=3
    
tree
```

```python
# right 1, down one
count = 0
tree = 0

for y in df.index:
       
    if df[count][y] == 1:
            
            tree +=1
            
    count+=1
    
tree, count
```

```python
# right 5, down 1
count = 0
verify = 0
tree = 0

for y in df.index:
    
    if count % 5 == 0:
        
        if df[count][y] == 1:
            
            tree +=1
            
    count+=5
    verify +=1
    
tree, verify
```

```python
# right 1, down 2
count = 0
tree = 0

for y in df.index:
        
    if y % 2 != 0:
        tree+=0
        
    elif y % 2 == 0:
        if df[count][y] == 1:
            tree +=1
        
        count+=1
    
tree, count
```
